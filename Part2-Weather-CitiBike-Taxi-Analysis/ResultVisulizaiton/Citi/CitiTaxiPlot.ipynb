```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline

#citi_taxi_dayofweek
df = pd.read_csv('/Users/Arafat/Desktop/citiData/taxi_citi_dayofweek_count.csv')
citi_taxi = df[['dayofweek','citi_dayofweek_count','taxi_dayofweek_count']] \
               .rename(columns={'citi_dayofweek_count':'The total number of citi used','taxi_dayofweek_count':'The total number of taxi used'})
ax = citi_taxi.plot(kind = 'bar',x='dayofweek',figsize=(12,6),secondary_y='The total number of citi used',colormap='winter',legend=True,linewidth=5.0)
ax.set_ylabel('The total number of taxi used')
ax.right_ax.set_ylabel('The total number of citi used')
ax.set_xlabel('Day Of Week')
ax.set_facecolor('purple')
plt.title('The trend of total number of pickups of yellow taxi and total number of Citi bike used from 2015 to 2016, By Dayofweek',fontsize=10.5)
plt.show()
```

```python
#citi_taxi_month_count
df = pd.read_csv('/Users/Arafat/Desktop/citiData/taxi_citi_month_count.csv')
citi_taxi = df[['month','citi_month_count','taxi_month_count']] \
               .rename(columns={'citi_month_count':'The total number of citi used','taxi_month_count':'The total number of taxi used'})
ax = citi_taxi.plot(kind = 'bar',x='month',figsize=(12,6),secondary_y='The total number of citi used',color=['blue','yellow'],legend=True)
ax.set_ylabel('The total number of taxi used')
ax.right_ax.set_ylabel('The total number of citi used')
ax.set_xlabel('Month')
ax.set_facecolor('pink')
plt.title('The trend of total number of pickups of yellow taxi and total number of Citi bike used from 2015 to 2016, By Month',fontsize=11)
plt.show()
```
