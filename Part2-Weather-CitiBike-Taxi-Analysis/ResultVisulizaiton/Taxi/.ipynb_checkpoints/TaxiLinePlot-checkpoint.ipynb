```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline

#taxi_pickup_hour_weekday
df = pd.read_csv('/Users/Arafat/Desktop/taxi/mon_to_sun_hour_amount.csv')
weekday_amount = df[['hour','mon_amount','tues_amount','wed_amount','thur_amount','fri_amount']] \
               .rename(columns={'mon_amount':'Monday','tues_amount':'Tuesday','wed_amount':'Wednesday','thur_amount':'Thursday','fri_amount':'Friday'})
ax = weekday_amount.plot(kind = 'line',x='hour',figsize=(12,6),xticks=df['hour'],linewidth=3.0,color=['deeppink','lightcoral','darkorange','steelblue','darkcyan'], legend=True)
ax.set_ylabel('The total income')
ax.set_xlabel('Hours')
ax.set_facecolor('peachpuff')
plt.title('The total income of yellow taxi in weekday(include Friday), By hours', fontsize=14)
plt.show()
```

```python
#taxi_pickup_hour_weekend
df = pd.read_csv('/Users/Arafat/Desktop/taxi/mon_to_sun_hour_amount.csv')
weekend_amount = df[['hour','fri_amount','sat_amount','sun_amount']] \
               .rename(columns={'sat_amount':'Saturday','sun_amount':'Sunday','fri_amount':'Friday'})
ax = weekend_amount.plot(kind = 'line',x='hour',figsize=(12,6),xticks=df['hour'],linewidth=4.0,color=['orangered','blueviolet','greenyellow'], legend=True)
ax.set_ylabel('The total income')
ax.set_xlabel('Hours')
ax.set_facecolor('cornflowerblue')
plt.title('The total income of yellow taxi in weekend(include Friday), By hours', fontsize=14)
plt.show()
```

```python
#shared ride vs days in 2015 and 2016
df = pd.read_csv('/Users/Arafat/Desktop/taxi/15ShareRidevsDay.csv')
shared_ride = df[['Date','sharedRideTimes2015','sharedRideTimes2016','singleRideTimes2015','singleRideTimes2016']] \
               .rename(columns={'sharedRideTimes2015':'Shared 2015','sharedRideTimes2016':'Shared 2016', \
                                'singleRideTimes2015':'Single 2015','singleRideTimes2016':'Single 2016'})
ax = shared_ride.plot(kind = 'line',x='Date',figsize=(12,6),linewidth=3.0,colormap='Set2', legend=True)
ax.set_ylabel('Number of shared ride')
ax.set_xlabel('Date')
ax.set_facecolor('thistle')
plt.title('# of shared and single ride times of yellow taxi in 2015 and 2016, by Date', fontsize=14)
plt.show()
```

```python
#shared ride vs hours in 2015
df = pd.read_csv('/Users/Arafat/Desktop/taxi/15ShareRidevsHour.csv')
shared_ride = df[['time','Monday2015','Tuesday2015','Wednesday2015','Thursday2015','Friday2015','Saturday2015','Sunday2015']] \
               .rename(columns={'Monday2015':'Monday','Tuesday2015':'Tuesday','Wednesday2015':'Wednesday','Thursday2015':'Thursday', \
                               'Friday2015':'Friday','Saturday2015':'Saturday','Sunday2015':'Sunday'})
ax = shared_ride.plot(kind = 'line',x='time',xticks=df['time'],figsize=(12,6),linewidth=3.0,colormap='winter', legend=True)
ax.set_ylabel('Number of shared ride')
ax.set_xlabel('Hours')
ax.set_facecolor('lightblue')
plt.title('# of shared ride times of yellow taxi in 2015, by Hour', fontsize=14)
plt.show()
```

```python
#shared ride vs hours in 2016
df = pd.read_csv('/Users/Arafat/Desktop/taxi/16ShareRidevsHour.csv')
shared_ride = df[['time','Monday2016','Tuesday2016','Wednesday2016','Thursday2016','Friday2016','Saturday2016','Sunday2016']] \
               .rename(columns={'Monday2016':'Monday','Tuesday2016':'Tuesday','Wednesday2016':'Wednesday','Thursday2016':'Thursday', \
                               'Friday2016':'Friday','Saturday2016':'Saturday','Sunday2016':'Sunday'})
ax = shared_ride.plot(kind = 'line',x='time',xticks=df['time'],figsize=(12,6),linewidth=3.0,colormap='Set1', legend=True)
ax.set_ylabel('Number of shared ride')
ax.set_xlabel('Hours')
ax.set_facecolor('mistyrose')
plt.title('# of shared ride times of yellow taxis in 2016, by Hour', fontsize=14)
plt.show()
```
