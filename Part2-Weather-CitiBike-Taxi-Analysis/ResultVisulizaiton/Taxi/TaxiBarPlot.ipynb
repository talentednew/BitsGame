```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline

#tip fraction vs number of passenger in 2015
df = pd.read_csv('/Users/Arafat/Desktop/taxi/TipFracvsPcnt15.csv')
tip_pger = df[['passenger_count','2015avg(tipFraction)','2016avg(tipFraction)']] \
               .rename(columns={'2015avg(tipFraction)':'2015','2016avg(tipFraction)':'2016'})
ax = tip_pger.plot(kind = 'bar',x='passenger_count',figsize=(12,5),color=['red','black'], legend=True,alpha = 0.5)
ax.set_ylabel('# of pickups')
ax.set_xlabel('# of passengers')
ax.set_facecolor('lightgreen')
plt.title('The relationship between tip fractions and number of passengers for 2015 and 2016')
plt.show()
```

```python
#single and shared ride vs rainy days in 2015
df = pd.read_csv('/Users/Arafat/Desktop/taxi/single16vsPRCP.csv')
sig_shar = df[['PRCP','single','shared']]
ax = sig_shar.plot(kind = 'bar',x='PRCP',figsize=(12,5),color=['lightblue','yellow'], legend=True,alpha = 0.7)
ax.set_ylabel('# of pickups')
ax.set_xlabel('Rain in inches')
ax.set_facecolor('lightpink')
plt.title('The relationship between rainy days and way of riding')
plt.show()
```
