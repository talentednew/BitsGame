```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline

#taxi_Rain_2015
df = pd.read_csv('/Users/Arafat/Desktop/WeatherData/taxi_rain_2015.csv')
rain_taxi = df[['Date','Rain','PickupTimes']] \
               .rename(columns={'PickupTimes':'Total number of pickups'})
ax = rain_taxi.plot(kind = 'bar',x='Date',figsize=(20,6),secondary_y='Rain',color=['dodgerblue','gold'],legend=True,linewidth=5.0)
ax.set_ylabel('The total number of taxi used')
ax.right_ax.set_ylabel('Rain in inches')
ax.set_xlabel('Date (2015)')
ax.set_facecolor('salmon')
plt.title('The trend of total number of pickups of yellow taxi and the Rain in inches, By Date in 2015',fontsize=15)
plt.show()
```

```python
#taxi_Rain_2016
df = pd.read_csv('/Users/Arafat/Desktop/WeatherData/taxi_rain_2016.csv')
rain_taxi = df[['Date','Rain','PickupTimes']] \
               .rename(columns={'PickupTimes':'Total number of pickups'})
ax = rain_taxi.plot(kind='bar',x='Date',figsize=(20,6),secondary_y='Rain',color=['dodgerblue','gold'],legend=True,linewidth=5.0,alpha=0.8)
ax.set_ylabel('The total number of taxi used')
ax.right_ax.set_ylabel('Rain in inches')
ax.set_xlabel('Date (2016)')
ax.set_facecolor('orangered')
plt.title('The trend of total number of pickups of yellow taxi and the Rain in inches, By Date in 2016',fontsize=15)
plt.show()
```

```python
#taxi_Snow_2015
df = pd.read_csv('/Users/Arafat/Desktop/WeatherData/taxi_snow_2015.csv')
snow_taxi = df[['Date','Snow','PickupTimes']] \
               .rename(columns={'PickupTimes':'Total number of pickups'})
ax = snow_taxi.plot(kind='bar',x='Date',figsize=(10,5),secondary_y='Snow',color=['blue','yellow'],legend=True,linewidth=5.0,alpha=0.5)
ax.set_ylabel('The total number of taxi used')
ax.right_ax.set_ylabel('Snow in inches')
ax.set_xlabel('Date (2015)')
ax.set_facecolor('lightskyblue')
plt.title('The trend of total number of pickups of yellow taxi and the Snow in inches, By Date in 2015',fontsize=12)
plt.show()
```
