```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline

#Weather_taxi_day_2015
df = pd.read_csv('/Users/Arafat/Desktop/WeatherData/weather_taxi_day_mean_2015.csv')
weather_taxi = df[['date','taxi_day_count_2015','TEMP']] \
               .rename(columns={'taxi_day_count_2015':'The total times of pickup','TEMP':'Average Temperature(F)'})
ax = weather_taxi.plot(kind = 'line',x='date',secondary_y='Average Temperature(F)',figsize=(12,5),color=['yellow','blue'], legend=True)
ax.set_ylabel('The total times of pickups')
ax.right_ax.set_ylabel('Average Temperature')
ax.set_xlabel('Date (2015)')
ax.set_facecolor('green')
plt.title('The trend of total number of pickups of yellow taxi and average temperature of NYC in 2015, By Date')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline

#weather_taxi_month_2015
df3 = pd.read_csv('/Users/Arafat/Desktop/WeatherData/weather_taxi_month_mean_2015.csv')
weather_taxi = df3[['month','taxi_month_count_2015','TEMP']] \
               .rename(columns={'taxi_day_count_2015':'The total times of pickup','TEMP':'Average Temperature(F)'})
ax = weather_taxi.plot(kind = 'line',x='month',secondary_y='Average Temperature(F)',figsize=(12,5),xticks=df3['month'],colormap='winter',legend=True,linewidth=5.0)
ax.set_ylabel('The total times of pickups')
ax.right_ax.set_ylabel('Average Temperature')
ax.set_xlabel('Month (2015)')
ax.set_facecolor('purple')
plt.title('The trend of total number of pickups of yellow taxi and average temperature of NYC in 2015, By Month',fontsize=11)
plt.show()
```
